```python
import matplotlib.pyplot as plt
# plot() 주어진 숫자값을 연결해서 선으로 만들어준다.
# x축과 y축 좌표를 지정하지 않을경우 x축은 0부터 1씩 증가
# y축은 사용자가 지정한 값으로 인식
plt.plot([1,4,5,8])
plt.show()
```

```python
import matplotlib.pyplot as plt
# x축과 y축 좌표를 지정한 경우
x = [2,4,6,8]
y = [1,3,5,7]
plt.plot(x,y)
plt.show()
```

```python
# 색상표
# b = 파란색
# g = 녹색
# r = 빨간색
# c = 청록색
# m = 마젠타색
# y = 노란색
# k = 검은색
# w = 흰색
# 마커
# o = 원
# v = 역삼각형
# ^ = 삼각형
# s = 사각형
# + = 플러스
# . = 점
x= range(0,10)
y = range(5,15)
plt.plot(x,y,'r+')
plt.show()
```
