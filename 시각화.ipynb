```python
import matplotlib.pyplot as plt
# 간격을 가능한 최적의 수치로 자동 조정
fig = plt.figure(constrained_layout = True)
# add_subplot(행의 갯수, 열의갯수, 위치값)
# 그래프의 위치 지정
ax1 = fig.add_subplot(1,2,1)
ax2 = fig.add_subplot(1,2,2)
# 원하는 위치에 그래프를 그림
x = range(0, 10)
y = range(5, 15)
ax1.plot(x, y, 'ro')
c = range(5, 50)
d = range(5, 50)
ax2.plot(c,d,'ms')
# x축이름, y축이름 지정
ax1.set_xlabel('time')
ax1.set_ylabel('qty')
ax2.set_xlabel('time')
ax2.set_ylabel('qty')
plt.show()
```

```python
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm
import matplotlib
# 간격을 가능한 최적의 수치로 자동 조정
fig = plt.figure(constrained_layout = True)
# 폰트 지정
font_location = "C:\Windows\Fonts\H2PORM.TTF"
fontname = fm.FontProperties(fname=font_location).get_name()
matplotlib.rc('font', family = fontname)
# add_subplot(행의 갯수, 열의갯수, 위치값)
# 그래프의 위치 지정
ax1 = fig.add_subplot(1,2,1)
ax2 = fig.add_subplot(1,2,2)
# 원하는 위치에 그래프를 그림
x = range(0, 10)
y = range(5, 15)
ax1.plot(x, y, 'ro')
c = range(5, 50)
d = range(5, 50)
ax2.plot(c,d,'ms')
# x축이름, y축이름 지정
ax1.set_xlabel('시간')
ax1.set_ylabel('수량')
ax2.set_xlabel('시간')
ax2.set_ylabel('수량')
plt.show()
```

```python
import matplotlib.pyplot as plt
# 막대그래프
plt.style.use('ggplot')
fig = plt.figure()
ax = fig.add_subplot(111)
x = [3, 4]
y = [1, 2]
# x축 항목 이름 지정
labels = ['english', 'math']
# 컬러 지정
colors = ['red', 'blue']
# bar() 막대그래프를 그리는 함수
ax.bar(x,y, tick_label = labels, color = colors)
for i, v in enumerate(y):
    ax.text(v+2, i+1.05,str(v),color = 'red', fontweight='bold')
plt.show()
```
