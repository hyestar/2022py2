```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
%matplotlib inline
sns.set_style("whitegrid")
iris = sns.load_dataset("iris")
x = iris.petal_length.values
sns.displot(x, rug=True)
plt.show()
```

```python
titanic = sns.load_dataset("titanic")
sns.countplot(x="class",data=titanic)
plt.show()
```

```python
tips = sns.load_dataset("tips")
sns.countplot(x="day",data=tips)
plt.show()
```

```python
tips.head(10)
```

```python
# 요일별 식사 금액
sns.boxplot(x="day",y ="total_bill",data=tips)
plt.show()
```

```python
# 요일별 식사 금액, 흡연자와 비흡연자를 비교
sns.boxplot(x="day",y ="total_bill",hue = "smoker",data=tips)
plt.show()
```

```python
# 요일별 식사 금액, 성별을 비교
sns.boxplot(x="day",y ="total_bill",hue = "sex",data=tips)
plt.show()
```

```python
sns.violinplot(x="day",y ="total_bill",hue = "sex",data=tips)
plt.show()
```

```python
sns.barplot(x="day",y ="total_bill",hue = "sex",data=tips)
plt.show()
```

```python
sns.set_style("darkgrid")
sns.lmplot(x ="total_bill", y = "tip", data=tips, size=7)
plt.show()
```

```python
# 식사 금액과 팁의 흡연자 비흡연자 기준, 색 변경
sns.lmplot(x ="total_bill", y = "tip", hue="smoker", data=tips, palette = "Set3", size=7)
plt.show()
```

```python
flights = sns.load_dataset("flights")
type(flights)
fgp=flights.pivot("month", "year", "passengers")
plt.figure(figsize=(10,8))
# YlGnBu 파란색과 노란색 계열
sns.heatmap(fgp, cmap="YlGnBu", annot = True, fmt = 'd')
plt.show()
```

```python
flights = sns.load_dataset("flights")
type(flights)
# pivot(인덱스, 컬럼, 데이터)
fgp=flights.pivot("month", "year", "passengers")
plt.figure(figsize=(10,8))
# YlGnBu 파란색과 노란색 계열
# RdYlGn_r
sns.heatmap(fgp, cmap="RdYlGn_r", annot = True, fmt = 'd')
plt.show()
```

```python
# stripplot 모든 데이터를 점으로 시각화
# jitter=True로 설정하면 위치를 무작위로 바꿔서 겹치지 않게 해줌
sns.stripplot(x="day", y='total_bill', data=tips, jitter=True)
plt.show()
```

```python
titanic = sns.load_dataset("titanic")
tit=titanic.pivot_table(index=["sex"], columns=["class"], aggfunc='size')
# linewidths 칸의 간격
sns.heatmap(tit, annot=True, fmt='d', linewidths=20)
plt.show()
```
